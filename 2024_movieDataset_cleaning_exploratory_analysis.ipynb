### Data Exploration --

```python
#find average movie Duration(minutes) and show Titles that are greater than the average
average_duration = newMovieData['Duration (minutes)'].mean()
average_duration
```

```python
#Show titles that are greater than the average movie duration
newMovieData[newMovieData['Duration (minutes)']>=120]['Title']
```

```python
# display top 10 lengthy movie
top10_length = newMovieData.nlargest(10, 'Duration (minutes)')[['Title','Duration (minutes)']]\
.set_index('Title')
# .set_index('Title') to ensure Title is used as index
top10_length
```

```python
# use of barplot to display a more visual representation of above

sns.barplot(x="Duration (minutes)", y=top10_length.index,data=top10_length)
plt.title("Top 10 Lengthy Movies")
plt.show

#y-axis is title, x-axis is duration(minutes)
```

#### Create New Column And Classify Movie Ratings [Excellent, Good, Average]

```python
# Create function to hold if- condition

def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
# Apply function defined above, add to dataframe as newMovieData['Rating_class']
newMovieData['Rating_class'] = newMovieData['IMDb Rating'].apply(rating)
```

```python
# Confirm new column 'Rating_class' have been added to dataset
newMovieData.head(5)
```

```python
# check for popuar Genre(By count) in entire dataset
newMovieData['Genre'].value_counts()
```

```python
# Top Genres by Rating
newMovieData.groupby('Genre')['IMDb Rating'].mean().sort_values(ascending=False)
```

```python
# barplot for Top Genres
sns.barplot(x='Genre',y='IMDb Rating',data=newMovieData)
plt.title("Top Rated Genres")
plt.xticks(rotation=45)
plt.show()
```
